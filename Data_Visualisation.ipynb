```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(0)
data = np.random.normal(50, 10, 100)

# Add some outliers to the data
outliers = np.random.normal(150, 5, 5)
data = np.concatenate((data, outliers), axis=0)

# Plot the data as a scatter plot
plt.scatter(range(len(data)), data)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.random.normal(0,1,100)
y=np.random.normal(0,1,100)

x_min,x_max=-1,1
y_min,y_max=-1,1

mask=(x>=x_min)&(x<=x_max)&(y>=y_min)&(y<=y_max)

plt.scatter(x,y,color="grey")

plt.scatter(x[mask],y[mask],color="black")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(0)
x=np.random.normal(0,1,50)
y=np.random.normal(0,1,50)

colors=[]
for i in range(len(x)):
  if x[i]>0 and y[i]>0:
    colors.append("Yellow")
  elif x[i]>0 and y[i]<0:
    colors.append("Blue")
  elif x[i]<0 and y[i]>0:
    colors.append("Green")
  else:
    colors.append("Red")
plt.scatter(x,y,color =colors)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate two sets of random data
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(2, 1, 1000)

# Set up the histogram parameters
bins = np.linspace(-5, 5, 50)

# Create the histograms
plt.hist(data1, bins, alpha=0.5, label='Data 1')
plt.hist(data2, bins, alpha=0.5, label='Data 2')
plt.legend(loc='upper right')

# Add axis labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Overlay')

# Show the plot
plt.show()
```
