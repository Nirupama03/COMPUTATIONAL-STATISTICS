Q2.2) CREATE A RADIAL CHART FOR THE DATA


```python
import numpy as np
import matplotlib.pyplot as plt


categories = ['Food Quality', 'Food Variety', 'Service Quality', 'Ambiance', 'Affordability']

restaurant_1 = [4, 4, 5, 4, 3]
restaurant_2 = [5, 5, 4, 5, 2]
restaurant_3 = [3, 4, 5, 3, 5]

label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(restaurant_1))

plt.figure(figsize=(8, 8))
plt.subplot(polar=True)
plt.plot(label_loc, restaurant_1, label='Restaurant 1')
plt.plot(label_loc, restaurant_2, label='Restaurant 2')
plt.plot(label_loc, restaurant_3, label='Restaurant 3')
plt.title('Restaurant comparison', size=20)
lines, labels = plt.thetagrids(np.degrees(label_loc), labels=categories)
plt.legend()
plt.show()
```

Q3) CREATE A CHART TABLE FOR THE DATA

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
data = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
rows = ['Row 1', 'Row 2']
columns = ['Column 1', 'Column 2', 'Column 3', 'Column 4']

# Create a table from the sample data
fig, ax = plt.subplots()
table = ax.table(cellText=data, rowLabels=rows, colLabels=columns, loc='center')
table.set_fontsize(14)
table.scale(1, 1.5)

# Remove the default axis
ax.axis('off')

# Show the plot
plt.show()
```


Q6)Draw a Scatter Plot for Four Quadrants differentiated by colors for the data.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
np.random.seed(0)
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

# Assign colors based on the quadrant in which the data points lie
colors = []
for i in range(len(x)):
    if x[i] > 0 and y[i] > 0:
        colors.append('red')
    elif x[i] < 0 and y[i] > 0:
        colors.append('blue')
    elif x[i] < 0 and y[i] < 0:
        colors.append('green')
    else:
        colors.append('yellow')

# Create a scatter plot from the sample data
plt.scatter(x, y, c=colors)

# Add labels and title to the plot
plt.title("Scatter Plot with Four Quadrants")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()

```

Q7) Draw a Scatter Plot for Outliers Highlighted for the data.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Generate some sample data
np.random.seed(0)
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

# Identify outliers using the Z-score method
z = np.abs(stats.zscore(np.column_stack((x, y))))
outliers = np.where(z > 3)

# Create a scatter plot from the sample data
plt.scatter(x, y, color='gray', alpha=0.6)

# Highlight the outliers in a different color
plt.scatter(x[outliers[0]], y[outliers[0]], color='red', alpha=1)

# Add labels and title to the plot
plt.title("Scatter Plot with Outliers Highlighted")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```

Q8)Draw a Scatter Plot for Areas Highlighted for the data.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
np.random.seed(0)
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

# Define the boundaries for the areas to be highlighted
x_min, x_max = -1, 1
y_min, y_max = -1, 1

# Identify the data points within the boundaries
mask = (x >= x_min) & (x <= x_max) & (y >= y_min) & (y <= y_max)

# Create a scatter plot from the sample data
plt.scatter(x, y, color='gray', alpha=0.6)

# Highlight the data points within the boundaries in a different color
plt.scatter(x[mask], y[mask], color='red', alpha=1)

# Add labels and title to the plot
plt.title("Scatter Plot with Areas Highlighted")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```
